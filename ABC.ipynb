```python
import numpy as np
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image
```

```python
print("heyfuy")
```

```python
TRAIN_PATH = 'seg_train/seg_train'
TEST_PATH = 'seg_test\seg_test'

train_gen = ImageDataGenerator(rescale=1./255).flow_from_directory(
    TRAIN_PATH, target_size=(200, 200), batch_size=32, class_mode='categorical')
test_gen = ImageDataGenerator(rescale=1./255).flow_from_directory(
    TEST_PATH, target_size=(200,200), batch_size=32, class_mode='categorical')
```

```python
num_classes = len(train_gen.class_indices)
class_name = list(train_gen.class_indices.keys())
print("number of clases", num_classes)
print("class name", class_name)
```

```python
model = Sequential(
    [
        # 32 is here no of filters by increasing it model can more complex features
        Conv2D(32, (3,3), activation='relu', input_shape=(200,200,3)),
        MaxPooling2D((2,2)),
        Conv2D(64, (3,3), activation='relu'),
        MaxPooling2D((2,2)),
        Conv2D(128, (3,3), activation='relu'),
        MaxPooling2D((2,2)),
        Conv2D(256, (3,3), activation='relu'),
        MaxPooling2D((2,2)),
        Flatten(),
        Dense(128, activation='relu'),
        Dropout(0.5),
        Dense(num_classes, activation='softmax')
    ]
                   
)
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(train_gen, epochs=20, validation_data=test_gen, batch_size =32)
```

```python
model.save('model.h5')
```
